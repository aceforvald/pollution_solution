```python
import pandas as pd
import json
import numpy as np
```

```python
# read json file
with open('iceland.json', 'r') as f:
    iceland_json = json.load(f)

iceland_json
```

```python
# slice data to have only sensor 10673
# put sensors name and parameters in different variables
sensor_10673_name = iceland_json["results"][0]["name"]
sensor_10673_parameters = iceland_json["results"][0]["parameters"]

sensor_10673_parameters
```

```python
# data as a pandas data frame
columns = ['location']
values = [sensor_10673_name]

for i in range(len(sensor_10673_parameters)):
    columns.append(sensor_10673_parameters[i]['displayName'])
    values.append(sensor_10673_parameters[i]['lastValue'])

df = pd.DataFrame(np.array(values).reshape(-1,len(values)), columns=columns)

# save data as a csv
df.to_csv("iceland.csv", index=False)

df
```
