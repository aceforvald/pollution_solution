```python
import requests
import json
from IPython.display import display, JSON

country_code = "US"

url = f"https://api.openaq.org/v2/locations?country={country_code}"
# https://api.openaq.org/v2/locations?country=me

headers = {"Accept": "application/json"}

response = requests.get(url, headers=headers)
```

```python
print(f'type response: {response}')
```

```python
data = response.json()
print(f'type: {type(data)}')

data_dumps = json.dumps(data, indent=4)
print('type: ', data_dumps)


#print(json.dumps(json_data, indent=4))

#data = json.loads(response)

#print(type(data))

#display(JSON(data_dumps))
```
