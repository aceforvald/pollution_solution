```python
lat = '40.42188'
lon = '-3.70079'
radius = '10000'

url = "https://api.openaq.org/v2/locations?limit=100&page=1&offset=0&sort=desc&coordinates=40.42188%2C-3.70079&radius=10000&order_by=lastUpdated&dump_raw=false"

sensors_response = requests.get(url, headers=headers)

sensors_data = sensors_response.json()

data_dumps = json.dumps(sensors_data, indent=4)

print('type: ', data_dumps)
```

lets look at that like a dataframe

```python
sensors_df = json_normalize(sensors_data['results'])
sensors_df = sensors_df.drop(columns=['entity', 'sources', 'isMobile', 'isAnalysis', 'sensorType', 'measurements', 'bounds', 'manufacturers'])
```

```python
#sensor_df['params'] = sensor_df['parameters']['parameter']
print(type(sensors_df['parameters']))
```

`parameters` contain list of dicts containing the specific parameters each sensor measures. We want to extract the type of sensor (`id` or `parameterId` - is there a difference?) maybe also the name (`parameter`) for readability.

```python
sensors_df['parameterIds'] = sensors_df['parameters'].apply(lambda x: [param_dict['id'] for param_dict in x])
sensors_df
```

```python
# 2, 1, 6, 19843, 3
sensors_df['has_pm25'] = sensors_df['parameterIds'].apply(lambda x: 1 if 2 in x else 0)
sensors_df['has_pm10'] = sensors_df['parameterIds'].apply(lambda x: 1 if 1 in x else 0)
sensors_df['has_so2'] = sensors_df['parameterIds'].apply(lambda x: 1 if 6 in x else 0)
sensors_df['has_no'] = sensors_df['parameterIds'].apply(lambda x: 1 if 19843 in x else 0)
sensors_df['has_o3'] = sensors_df['parameterIds'].apply(lambda x: 1 if 3 in x else 0)
sensors_df['sensorCount'] = sensors_df['has_pm25'] + sensors_df['has_pm10'] + sensors_df['has_so2'] + sensors_df['has_no'] + sensors_df['has_o3']
sensors_df
```

```python
places = {}

def add_place(city, country):
    OW_API = 'e1fb5f641995b86dc9040b6f5ed62f33'
    city_name = city
    country_code = country

    headers = {"Accept": "application/json",
            "X-API-Key": OW_API}

    url = f'http://api.openweathermap.org/geo/1.0/direct?q={city_name},{country_code}&limit=1&appid={OW_API}'
    response = requests.get(url, headers=headers)
    
    metro = response.json()

    places[metro[0]['name']] = [metro[0]['lat'], metro[0]['lon']]

    return places
```
